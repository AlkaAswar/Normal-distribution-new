```python
import pandas as pd
import numpy as np
```

```python
beml_df = pd.read_csv('/content/BEML.csv')
beml_df.head()
```

```python
glaxo_df = pd.read_csv('/content/GLAXO.csv')
glaxo_df[0:5]
```

```python
beml_df = beml_df[['Date','Close']]
glaxo_df = glaxo_df[['Date','Close']]
```

```python
beml_df.head()
```

```python
glaxo_df.head()
```

```python
''' The dataframes have a datecolumn, so we can create a datetimeIndex index from this column date. It will ensure that the rows are sorted by time is
ascending order.'''

glaxo_df = glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
glaxo_df
```

```python
beml_df = beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.plot(glaxo_df.Close)
plt.xlabel('Date')
plt.ylabel('Close Price')
```

```python
plt.plot(beml_df.Close)
plt.xlabel('Date')
plt.ylabel('Close Price')
```

```python
# (todays close price - pervious close price)/pervious close price
glaxo_df['Gain'] = glaxo_df.Close.pct_change(periods=1)
beml_df['Gain'] = beml_df.Close.pct_change(periods=1)
glaxo_df.head()
```

```python
glaxo_df = glaxo_df.dropna()
glaxo_df
```

```python
beml_df =beml_df.dropna()
```

```python
# Plots for gains
plt.figure(figsize = (8,6))
plt.plot(glaxo_df.index,glaxo_df.Gain)
plt.xlabel('Date')
plt.ylabel('Gain')
```

```python
# Plots for gains
plt.figure(figsize = (8,6))
plt.plot(beml_df.index,beml_df.Gain)
plt.xlabel('Date')
plt.ylabel('Gain')
```

```python
import seaborn as sns
sns.distplot(glaxo_df.Gain,label = 'Glaxo')
plt.xlabel('Gain')
plt.ylabel('Density')
```

```python
sns.distplot(beml_df.Gain,label = 'Beml')
plt.xlabel('Gain')
plt.ylabel('Density')
```

```python
print('Mean : ',round(glaxo_df.Gain.mean(),4))
print('Stanadared Deviation : ',round(glaxo_df.Gain.std(),4))
```

```python
print('Mean : ',round(beml_df.Gain.mean(),4))
print('Stanadared Deviation : ',round(beml_df.Gain.std(),4))
```

```python
from scipy import stats
# Probability of making 2% loss or higher in glaxo
stats.norm.cdf(-0.02,loc=glaxo_df.Gain.mean(),scale=glaxo_df.Gain.std()) * 100
```

```python
# Probability of making 2% loss or higher in Beml
stats.norm.cdf(-0.02,0.0003,0.0264 ) *100
```

```python
# Probability of making 2% gain or higher in Glaxo
df = 1-stats.norm.cdf(0.02,loc=glaxo_df.Gain.mean(),scale=glaxo_df.Gain.std())
df *100
```
